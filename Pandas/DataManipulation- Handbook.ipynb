```python
# Slicing with loc[], iloc
## DataFrames can be sliced by index values or by row/column
## .loc[] sliced by index values
### You can only slice an index if the index is sorted
## .iloc[] by row/column number
```

```python
temp_srt=dft.set_index(["country","city"]).sort_index()
temp_srt.head()
```

```python
temp_srt.loc["Pakistan":"Russia"]
```

```python
temp_srt.loc[("Pakistan","Lahore"):("Russia","Moscow")]
```

```python
temp_srt.loc[("India","Hyderbad"):("Iraq","Baghdad")]
```

```python
temp_srt.loc[:,"date":"avg_temp_c"]
```

```python
temp_srt.loc[("India", "Hyderabad"):("Iraq", "Baghdad"):,"date":"avg_temp_c"]
```
